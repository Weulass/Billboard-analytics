```python
#Python libraries we may need
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta
import numpy as np
```

```python
#upload the original database df
with open('hot-100_data.pkl', 'rb') as f:
        df = pickle.load(f)
```

```python
#we check the database
df
```

```python
# add features numberofappearances, best rank and first rank for each song
#we also count the number of songs each artist placed on the Hot 100 and Hot 10 before releasing the song 
df["numberofappearances"]=0
df["bestrank"]=0
df["firstrank"]=0
df["numberofappearances_artist"]=0
df["numberofappearances_artist_top10"]=0
for i in range(len(df)):
    title=df.title.loc[i]
    artist=df.artist.loc[i]
    date=df.date.loc[i]
    new_df=df[df["artist"] == artist][df["title"] == title][df["date"] <= date]
    table=new_df[new_df.artist==artist]
    table=table[table.title==title]
    df["numberofappearances"].loc[i]=len(table)
    df["bestrank"].loc[i]=table["rank"].min()
    df["firstrank"].loc[i]=table["rank"].iloc[-1]
        
```

```python
#creation of new_df, which aggregates df by artist

artists=df.artist.unique()
nb_songs=[]
new_df=pd.DataFrame()
new_df["artist"]=artists
for i in range(len(artists)):
    songs=len(df[df.artist==artists[i]].title.unique())
    nb_songs.append(songs)
    
```

```python
new_df["numberofsongs"]=nb_songs
```

```python
#number of top 10 reached by each artist

top=[]
for i in range(len(artists)):
    top_10=0
    song=df[df.artist==artists[i]].title.unique()
    for i in range(len(song)):
        if df[df.title==song[i]].bestrank.min() < 11:
            top_10=top_10+1
    top.append(top_10)
    
    
```

```python
#Best rank ever reached by each artist

bestrank=[]
for i in range(len(artists)):
    rank=df[df.artist==artists[i]].bestrank.min()
    bestrank.append(rank)
new_df["bestrankever"]=bestrank
```

```python
#number of top 50 reached by each artist

top_50=[]
for i in range(len(artists)):
    top=0
    song=df[df.artist==artists[i]].title.unique()
    for i in range(len(song)):
        if df[df.title==song[i]].bestrank.min() < 51:
            top=top+1
    top_50.append(top)
new_df["numberoftop50"]=top_50
```

```python
#number of top 75 reached by each artist

top_75=[]
for i in range(len(artists)):
    top=0
    song=df[df.artist==artists[i]].title.unique()
    for i in range(len(song)):
        if df[df.title==song[i]].bestrank.min() < 76:
            top=top+1
    top_75.append(top)
new_df["numberoftop75"]=top_75
```

```python
#number of top 25 reached by each artist

top_25=[]
for i in range(len(artists)):
    top=0
    song=df[df.artist==artists[i]].title.unique()
    for i in range(len(song)):
        if df[df.title==song[i]].bestrank.min() < 26:
            top=top+1
    top_25.append(top)
new_df["numberoftop25"]=top_25
```

```python
#number of weeks with at least one song in the Billboard Top 100

weeks=[]
for i in range(len(artists)):
    week=0
    week=len(df[df.artist==artists[i]].date.unique())
    weeks.append(week)
new_df["numberofweeks"]=weeks
```
