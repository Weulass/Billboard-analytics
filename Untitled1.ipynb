```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open('artist_df.pkl', 'rb') as f:
        artist_df = pickle.load(f)
```

```python
df= artist_df
```

```python
df
```

```python
df1 = df.drop([0],axis=1)  #delete first column, no information
df2 = df1.replace(to_replace='None', value=np.nan).dropna() 
```

```python
n_df = pd.DataFrame(columns=["artist","artists","award","award_date","format","label","last_certification","songs"])
l = df2['details'].tolist()
for i in range(len(df2)):
    certif = ['Platinum', 'Gold']
    new_l = []
    for j in range(len(l[i])):
        for word in certif:
            if word in l[i][j]:
                new_l.append(l[i][j])
                artist = df2.iloc[i][0]
                artists = df2.iloc[i][1]
                award = l[i][j].split("|")[0]
                award_date = l[i][j].split("|")[1]
                label = df2.iloc[i][4]
                formats = df2.iloc[i][3]
                last_certif = df2.iloc[i][5]
                songs = df2.iloc[i][6]
                n_df.loc[len(n_df)] = [artist,artists,award,award_date,formats,label,last_certif,songs]
```

```python
n_df
```

```python
pickle_out = open("artist.pickle","wb")
pickle.dump(n_df, pickle_out)
pickle_out.close()
```
